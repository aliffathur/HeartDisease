```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('heart.csv')

# Tampilkan lima baris pertama dari dataset
df.head()
```

```python
sns.countplot(x='target', data=df)
plt.title('Distribusi Penyakit Jantung')
plt.xlabel('0: Tidak Ada Penyakit Jantung, 1: Ada Penyakit Jantung')
plt.ylabel('Jumlah Pasien')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Korelasi Antar Fitur')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='target', kde=True, bins=30)
plt.title('Distribusi Usia Berdasarkan Penyakit Jantung')
plt.xlabel('Usia')
plt.ylabel('Jumlah Pasien')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Mengambil fitur yang relevan
X = df[['age', 'trestbps', 'chol', 'thalach', 'oldpeak']]

# Mengelompokkan data menjadi 3 cluster
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
df['cluster'] = kmeans.labels_

# Visualisasi cluster berdasarkan usia dan kolesterol
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='chol', hue='cluster', palette='Set2')
plt.title('Cluster Berdasarkan Usia dan Kolesterol')
plt.xlabel('Usia')
plt.ylabel('Kolesterol')
plt.show()
```
